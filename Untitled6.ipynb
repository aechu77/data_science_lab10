### 3. Data Filtering

```python
# Filter for 'setosa' species
setosa_df = df[df['species'] == 'setosa']
print("\nDataFrame filtered for 'setosa' species (first 5 rows):")
display(setosa_df.head())

# Filter for observations where 'sepal length (cm)' is greater than 5.0
long_sepal_df = df[df['sepal length (cm)'] > 5.0]
print("\nDataFrame filtered for 'sepal length (cm)' > 5.0 (first 5 rows):")
display(long_sepal_df.head())

# Filter for 'versicolor' species with 'petal length (cm)' greater than 4.0
versicolor_long_petal_df = df[(df['species'] == 'versicolor') & (df['petal length (cm)'] > 4.0)]
print("\nDataFrame filtered for 'versicolor' species and 'petal length (cm)' > 4.0 (first 5 rows):")
display(versicolor_long_petal_df.head())
```

### 4. Grouping and Aggregate Statistics

```python
# Group by 'species' and calculate the mean of all numerical features
print("\nMean of features grouped by species:")
display(df.groupby('species').mean())

# Group by 'species' and calculate descriptive statistics for 'sepal length (cm)'
print("\nDescriptive statistics for 'sepal length (cm)' grouped by species:")
display(df.groupby('species')['sepal length (cm)'].describe())

# Group by 'species' and find the maximum 'petal width (cm)'
print("\nMaximum 'petal width (cm)' grouped by species:")
display(df.groupby('species')['petal width (cm)'].max())
```

### 1. Import Libraries and Load Dataset

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Map numerical species to actual names for better readability
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species'] = df['species'].map(species_map)

# Display the first 5 rows of the dataset
print("First 5 rows of the Iris dataset:")
display(df.head())
```

### 2. Basic Data Exploration

```python
# Get general information about the DataFrame
print("\nDataFrame Info:")
df.info()

# Get descriptive statistics of the numerical features
print("\nDescriptive statistics:")
display(df.describe())
```
