```python
import numpy as np

# 3. Simulate a coin toss experiment and compute probabilities
num_tosses = 1000 # Number of coin tosses

# Simulate coin tosses (0 for tails, 1 for heads)
tosses = np.random.choice([0, 1], size=num_tosses)

# Count heads and tails
num_heads = np.sum(tosses)
num_tails = num_tosses - num_heads

# Compute probabilities
prob_heads = num_heads / num_tosses
prob_tails = num_tails / num_tosses

print(f"Number of tosses: {num_tosses}")
print(f"Number of heads: {num_heads}")
print(f"Number of tails: {num_tails}")
print(f"Probability of heads: {prob_heads:.4f}")
print(f"Probability of tails: {prob_tails:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate a normal distribution
# Parameters for the normal distribution
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000) # generate 1000 samples
```
