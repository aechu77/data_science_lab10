```python
def factorial(n):
    if n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res

# Example usage:
number = 5
print(f"The factorial of {number} is {factorial(number)}")
```

# Task
Create an expert system in Python that includes a knowledge base, an inference engine, and a user interface to make decisions based on a defined set of rules.

## Define Decision Rules

### Subtask:
Establish a simple set of 'if-then' rules for a decision-making scenario (e.g., what to wear based on weather) that the expert system will use.


## Defined Decision Rules

### Scenario: Outfit Recommendation Based on Weather

This expert system will recommend an outfit based on the current temperature and weather conditions.

### If-Then Rules:

1.  **IF** temperature is `below 10°C` **AND** weather is `rainy` **THEN** recommend `Heavy Coat, Waterproof Boots, Umbrella`.
2.  **IF** temperature is `between 10°C and 20°C` **AND** weather is `cloudy` **THEN** recommend `Light Jacket, Jeans, Closed Shoes`.
3.  **IF** temperature is `above 20°C` **AND** weather is `sunny` **THEN** recommend `T-shirt, Shorts, Sandals, Sunglasses`.
4.  **IF** temperature is `between 10°C and 20°C` **AND** weather is `sunny` **THEN** recommend `Light Sweater, Pants, Sneakers`.
5.  **IF** temperature is `below 10°C` **AND** weather is `snowy` **THEN** recommend `Winter Coat, Scarf, Gloves, Snow Boots`.

## Implement Knowledge Base

### Subtask:
Create a Python data structure (e.g., a list of dictionaries or a class) to store the defined rules, acting as the expert system's knowledge base.


**Reasoning**:
The subtask requires creating a Python list of dictionaries to serve as a knowledge base for an expert system. I will define each rule as a dictionary with 'conditions' and 'recommendation' keys and then append them to the `knowledge_base` list.



```python
knowledge_base = []

# Rule 1: Below 10°C, rainy -> Heavy coat, umbrella, waterproof boots
knowledge_base.append({
    'conditions': {'min_temp': None, 'max_temp': 9.9, 'weather': 'rainy'},
    'recommendation': 'Heavy coat, umbrella, waterproof boots'
})

# Rule 2: Between 10°C and 20°C, cloudy -> Light jacket, long pants, comfortable shoes
knowledge_base.append({
    'conditions': {'min_temp': 10.0, 'max_temp': 20.0, 'weather': 'cloudy'},
    'recommendation': 'Light jacket, long pants, comfortable shoes'
})

# Rule 3: Above 20°C, sunny -> T-shirt, shorts, sandals
knowledge_base.append({
    'conditions': {'min_temp': 20.1, 'max_temp': None, 'weather': 'sunny'},
    'recommendation': 'T-shirt, shorts, sandals'
})

# Rule 4: Below 10°C, snowy -> Winter coat, gloves, hat, snow boots
knowledge_base.append({
    'conditions': {'min_temp': None, 'max_temp': 9.9, 'weather': 'snowy'},
    'recommendation': 'Winter coat, gloves, hat, snow boots'
})

# Rule 5: Between 10°C and 20°C, sunny -> Light sweater, jeans, sneakers
knowledge_base.append({
    'conditions': {'min_temp': 10.0, 'max_temp': 20.0, 'weather': 'sunny'},
    'recommendation': 'Light sweater, jeans, sneakers'
})

print(knowledge_base)
```

## Develop Inference Engine

### Subtask:
Write a Python function or class that takes user inputs (e.g., temperature, weather condition) and applies the rules from the knowledge base to derive a decision or recommendation.


**Reasoning**:
I need to implement the `infer_outfit` function as described, incorporating the logic to match temperature and weather conditions against the rules in the `knowledge_base`, and then test it with example inputs.



```python
def infer_outfit(temperature, weather):
    """
    Infers an outfit recommendation based on temperature and weather conditions.
    """
    for rule in knowledge_base:
        conditions = rule['conditions']
        recommendation = rule['recommendation']

        # Check temperature conditions
        temp_match = False
        min_temp = conditions.get('min_temp')
        max_temp = conditions.get('max_temp')

        if min_temp is None and max_temp is None:
            temp_match = True
        elif min_temp is None and temperature <= max_temp:
            temp_match = True
        elif max_temp is None and temperature >= min_temp:
            temp_match = True
        elif min_temp <= temperature <= max_temp:
            temp_match = True

        # Check weather condition
        weather_match = (conditions.get('weather') == weather.lower())

        # If both conditions match, return the recommendation
        if temp_match and weather_match:
            return recommendation

    # If no rule matches
    return 'No specific recommendation available for these conditions. Stay comfortable!'

# Test the inference engine
print(f"Recommendation for 5°C and rainy: {infer_outfit(5, 'rainy')}")
print(f"Recommendation for 25°C and sunny: {infer_outfit(25, 'sunny')}")
print(f"Recommendation for 15°C and cloudy: {infer_outfit(15, 'cloudy')}")
print(f"Recommendation for 8°C and snowy: {infer_outfit(8, 'snowy')}")
print(f"Recommendation for 18°C and sunny: {infer_outfit(18, 'sunny')}")
print(f"Recommendation for 22°C and cloudy: {infer_outfit(22, 'cloudy')}") # No match expected
```

**Reasoning**:
The previous code produced a TypeError because the temperature comparison logic did not correctly handle cases where `min_temp` or `max_temp` could be `None` when attempting to compare them. I will refine the `temp_match` logic to explicitly check for `None` values for `min_temp` and `max_temp` before performing numerical comparisons, ensuring that comparisons are only made between numeric types.



```python
def infer_outfit(temperature, weather):
    """
    Infers an outfit recommendation based on temperature and weather conditions.
    """
    for rule in knowledge_base:
        conditions = rule['conditions']
        recommendation = rule['recommendation']

        # Check temperature conditions
        temp_match = False
        min_temp = conditions.get('min_temp')
        max_temp = conditions.get('max_temp')

        # Determine if temperature matches the rule's conditions
        if min_temp is not None and max_temp is not None:
            if min_temp <= temperature <= max_temp:
                temp_match = True
        elif min_temp is not None and max_temp is None:
            if temperature >= min_temp:
                temp_match = True
        elif min_temp is None and max_temp is not None:
            if temperature <= max_temp:
                temp_match = True
        elif min_temp is None and max_temp is None:
            # If neither min_temp nor max_temp is specified, it matches all temperatures
            # (though this case isn't strictly used in the current rule set)
            temp_match = True

        # Check weather condition
        weather_match = (conditions.get('weather') == weather.lower())

        # If both conditions match, return the recommendation
        if temp_match and weather_match:
            return recommendation

    # If no rule matches
    return 'No specific recommendation available for these conditions. Stay comfortable!'

# Test the inference engine
print(f"Recommendation for 5°C and rainy: {infer_outfit(5, 'rainy')}")
print(f"Recommendation for 25°C and sunny: {infer_outfit(25, 'sunny')}")
print(f"Recommendation for 15°C and cloudy: {infer_outfit(15, 'cloudy')}")
print(f"Recommendation for 8°C and snowy: {infer_outfit(8, 'snowy')}")
print(f"Recommendation for 18°C and sunny: {infer_outfit(18, 'sunny')}")
print(f"Recommendation for 22°C and cloudy: {infer_outfit(22, 'cloudy')}") # No match expected
```

## Create User Interface

### Subtask:
Implement a simple interactive loop that prompts the user for information and displays the decision provided by the inference engine.


**Reasoning**:
I will implement the interactive loop for the user interface, including input prompts, error handling for temperature, calling the inference engine, displaying the recommendation, and providing an option to continue or exit, as per the instructions.



```python
print("Welcome to the Outfit Recommender Expert System!")

while True:
    # 2. Prompt for temperature with error handling
    while True:
        try:
            temp_input = input("Enter the current temperature in Celsius (e.g., 15): ")
            temperature = float(temp_input)
            break
        except ValueError:
            print("Invalid temperature. Please enter a number.")

    # 3. Prompt for weather condition
    weather = input("Enter the current weather condition (e.g., sunny, rainy, cloudy, snowy): ").lower()

    # 4. Call the infer_outfit function
    recommendation = infer_outfit(temperature, weather)

    # 5. Print the recommendation
    print(f"Our recommendation: {recommendation}")

    # 6. Ask to continue or break the loop
    another_recommendation = input("Do you want another recommendation? (yes/no): ").lower()
    if another_recommendation == 'no':
        print("Thank you for using the Outfit Recommender!")
        break
```
